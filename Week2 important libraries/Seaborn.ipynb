```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
a=pd.read_csv("flights.csv")
print(a)
```

```python
a=sns.load_dataset("flights")
print(a)
```

```python
sns.relplot(x="passengers",y="month",hue="year",data=a)
```

```python
x=sns.load_dataset('tips')
```

```python
print(x)
```

```python
sns.relplot(x="time",y="tip",hue="day",data=x,kind='line')
```

```python
sns.catplot(x="day",y="total_bill",data=x,kind='boxen')
```

```python
sns.catplot(x="day",y="total_bill",data=x,kind='violin')
```
