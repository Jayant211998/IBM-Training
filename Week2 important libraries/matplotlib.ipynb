# Matplotlib

```python
from matplotlib import pyplot as plt
from matplotlib import style 
```

```python
plt.plot([1,2,3,6],[2,5,7,4])
plt.show()
```

```python
x=[1,5,4]
y=[23,30,40]

x2=[3,7,3]
y2=[5,8,9]

plt.plot(x,y)
plt.plot(x2,y2)

plt.show()
```

```python
style.use('ggplot')
x=[1,5,4]
y=[23,30,40]

x2=[3,7,3]
y2=[5,8,9]

plt.plot(x,y,label='first')
plt.plot(x2,y2,label='second')
plt.legend()
plt.show()
```

```python
style.use('ggplot')
x=[1,5,4]
y=[23,30,40]

x2=[3,7,3]
y2=[5,8,9]

plt.plot(x,y,label='first',linewidth=10)
plt.plot(x2,y2,label='second',linewidth=5)
plt.legend()
plt.show()
```

```python
plt.bar(x,y)
plt.bar(x2,y2)
plt.show()
```

```python
x=[1,5,4]
y=[23,30,40]
x2=[3,7,3]
y2=[5,8,9]

plt.hist(x,y)
plt.hist(x2,y2)
plt.show()
```

```python
plt.scatter(x,y,linewidth=3)
plt.scatter(x2,y2,linewidth=3)
plt.show()
```
